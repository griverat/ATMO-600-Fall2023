```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams["font.family"] = "monospace"
```

```python
non_dimy = np.linspace(-3.5, 3.5, 100)
non_dimx = np.linspace(-np.pi * 2, np.pi * 2, 100)
x, y = np.meshgrid(non_dimx, non_dimy)

k = 1
w = 1
t = np.pi / 2
# Create the wave function
u = np.exp(-(y**2) / 2) * np.exp(1j * (k * x - w * t))
# Create the convergence function
v = np.exp(-(y**2) / 2) * np.exp(1j * (k * x - w * t)) * 1j
```

```python
fig, ax = plt.subplots(figsize=(10, 5), dpi=300)
ax.contour(
    x,
    y,
    u.real,
    levels=np.concatenate([np.arange(-1.1, 0, 0.2), np.arange(0.1, 1, 0.2)]),
    colors="k",
    linewidths=1.2,
)

# plot the quiver with spacing 10
spacing = 3

for n in [-1, 3]:
    ax.text(
        n * np.pi / 2,
        0,
        "L",
        fontsize=25,
        color="Blue",
        ha="center",
        va="center",
        weight="bold",
    )

for n in [-3, 1]:
    ax.text(
        n * np.pi / 2,
        0,
        "H",
        fontsize=25,
        color="red",
        ha="center",
        va="center",
        weight="bold",
    )

# plot the divergence/convergence
v_ = np.where(abs(v.real) > 0.1, v, np.nan)
ax.contourf(
    x, y, v_.real, levels=np.arange(-1.1, 1.2, 0.2), cmap="coolwarm_r", alpha=0.5
)

# plot the arrows
u_ = np.where(abs(u.real) > 0.1, u, np.nan)
ax.quiver(
    x[::spacing, ::spacing],
    y[::spacing, ::spacing],
    u_.real[::spacing, ::spacing],
    0,
    pivot="middle",
    scale=50,
    color="green",
)

# turn off ticks
ax.set_xticks([])
# ax.set_yticks([])

ax.set_title("Theoretical equatorially trapped Kelvin wave")
ax.text(
    0,
    -0.01,
    r"(arrows) zonal wind field $u$; (contours) geopotential $\phi$"
    "\n(shading) convergence (red) and divergence (blue)",
    ha="left",
    va="top",
    transform=ax.transAxes,
    fontsize=8,
)

ax.text(
    1,
    -0.01,
    "Gerardo A. Rivera Tello\nATMO600 HW3",
    ha="right",
    va="top",
    transform=ax.transAxes,
    fontsize=7,
)
```
